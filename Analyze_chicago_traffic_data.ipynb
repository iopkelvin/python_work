```python
import pandas as pd
import numpy as np

%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
traffic_data = pd.read_csv("http://s3.amazonaws.com/gamma-datasets/Speed_Camera_Violations.csv")
```

```python
traffic_data.columns = traffic_data.columns.str.lower().str.replace(" ", "_")
```

```python
#Formatting datetime
traffic_data['violation_date'] = pd.to_datetime(traffic_data.violation_date)
```

```python
#Formatting day of the week
traffic_data['day_of_week'] = traffic_data.violation_date.dt.dayofweek
```

```python
traffic_data.sort_values(by='violation_date')
```

```python
# Question 1: What is the average number of citations issued per camera, per day?
average_violations_per_day = traffic_data.groupby("day_of_week")['violations'].mean()
average_violations_per_camera = traffic_data.groupby("camera_id")['violations'].mean()
print("Average amount of violations per day of week", average_violations_per_day, '\n')
print("Average amount of violations per Camera ID", average_violations_per_camera)
```

```python
traffic_data.camera_id.nunique()
```

```python
# Question 2: On which day of the week are the most citations issued
print("Average amount of citations per day of week", traffic_data.groupby("day_of_week")['violations'].mean())
print("Highest average of citations is observed on day 5, Saturday")
#Most citations are given on Saturday
```

```python
# Question 3: Have the number of active cameras increased or decreased over the collection period?
citationid_per_date = traffic_data[['violation_date', 'camera_id']].sort_values(by = "violation_date")
citationid_per_date.camera_id.nunique()
active_camera_per_date = citationid_per_date.groupby("violation_date")['camera_id'].count()
print("Although there is a high range of changes in activeID's during the years; The amount of active camera ID's has slightly increased over the years")
active_camera_per_date.plot()
plt.title('active IDs over time')
plt.xlabel("range in time")
plt.ylabel("Number of active camera IDs")
```

```python
print("Amount of active cameras per given day: \n", active_camera_per_date)
```

```python
# Question 4: Are there any interesting patterns or specific outliers you find in the data?
print(traffic_data.describe())
sns.heatmap(traffic_data.corr())
```

```python
print("outlier 1:\n There were only 16 active cameras during february 1st of 2015:\n", active_camera_per_date[active_camera_per_date == 16])
```

```python
traffic_data.groupby("violation_date")["violations"].mean().plot()
plt.title("Amount of citations by date")
plt.ylabel("Amount of citations")
plt.xlabel("Date")
print("The amount of citations given has been decreasing over the years, and also: During the first and last months of the year, the amount of citations given is very low. (Outlier)")
```

```python
traffic_nonnull = traffic_data.dropna()
print("Observations in the dataset")
print("non-null observations: ", traffic_nonnull.shape[0])
print("all observations: ", traffic_data.shape[0])
#Percentage of examples without null 
print("Percentage of non null observations: ", traffic_nonnull.shape[0]/ traffic_data.shape[0])
print("The dataset can further be cleaned by taking out the rows with null variables. The dataset without null variables accounts for 96.1% of the entire dataset")
```
